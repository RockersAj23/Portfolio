```python
from matplotlib import pyplot as plt
import numpy as np
plt.style.use('fivethirtyeight')
import pandas as pd
data = pd.read_csv(r"C:\Users\Ajay Mulgir\Downloads\data.csv")

# plt.xlabel('age')
# plt.ylabel('mean salary')
# plt.title("ye lo dekh lo")
# plt.show()
```

```python
data.head()
```

```python
from collections import Counter
language_counter=Counter()

for row in data['LanguagesWorkedWith']:
    language_counter.update(row.split(';'))

#print(language_counter.most_common(15))

lang=[]
popularity=[]
for item in language_counter.most_common(15):
    lang.append(item[0])
    popularity.append(item[1])

print(lang)
print(popularity)
```

```python
plt.bar(lang,popularity)
plt.xlabel('language')
plt.ylabel('no of persons know')
plt.title("we are booooming....")
plt.tight_layout()
plt.show()
```

```python
plt.barh(lang,popularity)
plt.xlabel('no of persons know')
plt.title("we are booooming....")
plt.tight_layout()
plt.show()
```

```python
# i want java on top
lang.reverse()
popularity.reverse()
plt.barh(lang,popularity)
plt.xlabel('no of persons know')
plt.title("we are booooming....")
plt.tight_layout()
plt.show()
```
