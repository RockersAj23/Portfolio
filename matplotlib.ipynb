```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=[11,10,12,14,12,15,14,13,15,102,12,14,17,19,107,10,13,12,14,12,108,12,11,14,13,15,10,15,12,10,14,13,15,10]
```

```python
plt.hist(dataset)
```

```python
#z-score

def outlier_finder(data):
    threshold=3
    outliers=[]
    mean = np.mean(data)
    std=np.std(data)

    for i in data:
        value=(i-mean)/std
        if value > threshold:
            outliers.append(i)

    return outliers
    
```

```python
outlier_finder(dataset)
```

```python
##calculating outlier by IQR method
q1,q3=np.percentile(dataset,[25,75])
```

```python
iqr=q3-q1
lf=q1-(1.5*q1)
uf=q3+(1.5*q3)
for i in dataset:
    if i< lf or i>uf:
        print(i)
```

```python
import pandas as pd
data  = pd.read_csv(r"C:\Users\Ajay Mulgir\Downloads\data.csv")
data.head()
```

```python
from collections import Counter
lang_counter=Counter()
for p in data['LanguagesWorkedWith']:
    lang_counter.update(p.split(';'))
lang_counter
```

```python
languagus=[]
popularity=[]
for item in lang_counter.most_common(15):
    languagus.append(item[0])
    popularity.append(item[1])

print(languagus)
print(popularity)
```

```python
plt.bar(languagus,popularity,color='#008fd5')


plt.tight_layout()
plt.xlabel('name of languages')
plt.ylabel('popularity')
plt.title('popularity of languages')
plt.show()
# as you can see if there are more bars , x axis isnt that much readble
```

```python
plt.barh(languagus,popularity,color='#008fd5')# barh is for horizental bars
plt.tight_layout()
plt.xlabel('name of languages')
plt.ylabel('popularity')
plt.title('popularity of languages')
plt.show()
#you notice that, if most popular at the top will be more likely
```

```python
languagus.reverse()
popularity.reverse()
width=0.4
plt.barh(languagus,popularity,color='#008fd5')# barh is for horizental bars
plt.tight_layout()
plt.xlabel('no. of people know')
plt.title('popularity of languages')
plt.show()
#here name of lang is self explaining this remove y_label
```

```python
import random 
from itertools import count
import pandas as pd 
import matplotlib.pyplot as plt 
from matplotlib.animation import FuncAnimation
plt.style.use('fivethirtyeight')
x_values=[]
y_values=[]
index= count()

def animate(i):
    x_values.append(next(index))
    y_values.append(random.randint(0,5))
    plt.cla()
    plt.plot(x_values,y_values)

anim=FuncAnimation(plt.gcf(),animate,interval=2000,cache_frame_data=False)
plt.tight_layout()
plt.show()
```

```python

import random
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []

index = count()


def animate(i):
    data = pd.read_csv('data.csv')
    x = data['x_value']
    y1 = data['total_1']
    y2 = data['total_2']

    plt.cla()

    plt.plot(x, y1, label='Channel 1')
    plt.plot(x, y2, label='Channel 2')

    plt.legend(loc='upper left')
    plt.tight_layout()


ani = FuncAnimation(plt.gcf(), animate, interval=1000)

plt.tight_layout()
plt.show()
```
